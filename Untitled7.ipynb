```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df=pd.read_csv('datasets/Pokemon.csv')
df.head(10)
```

```python
df.columns
```

```python
df.drop(['Name2'],
        axis=1,
        inplace=True
       )
df.columns
```

```python
pivot=pd.pivot_table(df,
               index='Primary Type',
                       values='Attack',
                       aggfunc='mean')
pivot
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

```python
ataque=pivot.sort_values(by='Attack',
                            ascending=False,
                            axis=0
                           )
ataque.head(5)
```

```python
barras=ataque.head(5)
barras.plot(kind='barh',
             figsize=(20,10),
             color='r',
             alpha=0.9)
```

```python
stats=list(df.loc[:,'Attack':'Speed'])
stats
```

```python
df.set_index('Name',
             inplace=True)
df.head(2)
```

```python
comparar=df.loc[['Entei', 'Lugia', 'Dialga'], stats]
comparar.plot(kind='line',
             figsize=(20,10),
              alpha=0.9
             )
```
