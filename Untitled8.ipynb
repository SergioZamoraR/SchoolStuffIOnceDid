```python
df=pd.read_excel('datasets/mic.xlsx',
                engine= 'openpyxl')
df.head()
```

```python
counts=df['Age'].value_counts()
counts
```

```python
counts=counts.to_dataframe()
counts
```

```python
counts
```

```python
counts['Total%']=(counts.loc[:,'Age'])/counts.loc[:,'Age'].sum(axis=0)*100
counts
```

```python
counts.reset_index(inplace=True)
counts.rename({'index':'Age_group'},
               axis = 1,
               inplace = True)
counts
```

```python
counts.drop(['level_0'],
            axis=1,
            inplace=True)
counts
```

```python
x=list(counts['Age_group'])
y=list(counts['Total%'])
```

```python
type(x)
mpl.style.use('ggplot')
```

```python
plt.subplot(1,2,1)
plt.rcParams['figure.figsize']=(15,7)
plt.bar(x,y,
       width=0.5,
       color='cyan')
plt.xlabel('Age group', fontweight='regular', color='black',
          fontsize=17, horizontalalignment='center')
plt.ylabel('Total %',fontweight='regular', color='black',
          fontsize=17, horizontalalignment='center')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.title('Porcentaje de encuestados de diferente categoria de edad')

plt.subplot(1,2,2)
plt.rcParams['figure.figsize']=(15,7)
fav_colors=['gold', 'yellowgreen', 'coral', 'skyblue', 'pink', 'strongorange']
exp_distance=[0.03, 0.05,0.02,0.04]
counts['Total%'].plot(kind='pie',
                      figsize=(15,6),
                      autopct='%1.1f%%',
                      colors=fav_colors,
                      startangle=90,
                      pctdistance=1.1,
                      labels=None,
                      explode=exp_distance,
                      shadow=True
                     )
plt.axis('equal')
plt.legend(labels=counts.Age_group, bbox_to_anchor=(1.0,1.0))
plt.title('Porcentaje de encuestados de diferente categoria de edad', y=1.02)
plt.savefig('age_group.png')
plt.show()
```
