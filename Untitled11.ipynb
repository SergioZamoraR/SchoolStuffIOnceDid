```python
import statsmodels.api as sm
```

```python
df=pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head(2)
```

```python
model=sm.OLS.from_formula('BPXSY1 ~ RIDAGEYR', data=df)
result=model.fit()
result.summary()
```

```python
ax=sns.regplot(x='BPXSY1',
               y='RIDAGEYR',
               data=df,
               x_estimator=np.mean
              )
```

```python
model=sm.OLS.from_formula('BMXHT~RIAGENDR+ BPXSY1+BMXWT',
                          data=df)
result=model.fit()
result.summary()
```

```python
x=result.fittedvalues
y=result.resid
```

```python
fig=sns.scatterplot(x=x,
                    y=y)
fig.set_xlabel('Fittedvalues')
fig.set_ylabel('Residuales')
```
