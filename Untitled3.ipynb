```python
print=np.pi
x=np.arange(0,3*np.pi,0.01)
y=np.sin(x)
plt.plot(x,y)
plt.show
y
```

```python
ysin=np.sin(x)
xcos=np.cos(x)
plt.plot(x,ysin)
plt.plot(x,xcos)
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.title('Sin and Cos')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,ysin)
plt.title('Sine')
plt.subplot(2,2,2)
plt.plot(x,xcos)
plt.title('Cosine')
plt.subplot(2,2,3)
plt.plot(x,ysin)
plt.title('Sine')
plt.subplot(2,2,4)
plt.plot(x,xcos)
plt.title('Cosine')
plt.show()
```

```python
import seaborn as sns
```

```python
path='datasets/Cartwheeldata.csv'
df2=pd.read_csv(path)
df2
```

```python
sns.lmplot(x='Wingspan',
           y='Height',
           data=df2,
           fit_reg=False,
           hue='Gender'
          )
plt.title('Scatterplot of Wingspan vs Height')
plt.show()
```

```python
sns.swarmplot(x='Gender',
              y='Wingspan',
              data=df2
             )
plt.title('Gender vs Wingspan, the Swarmplot')
plt.show()
```

```python
sns.boxplot(data=df2.loc[df2['Gender']=='F',
                         ['Age', 'Height', 'Wingspan', 'CWDistance']])
plt.title('Female Boxplot')
```

```python
sns.distplot(df2.CWDistance,
            kde=False)
```

```python
sns.countplot(x='Gender', 
              data=df2)
plt.xticks(rotation=180)
plt.yticks(rotation=45)
plt.show()
```
