```python
import statsmodels.api as sm
```

```python
df=pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head(2)
```

```python
df["RIAGENDRx"] = df.RIAGENDR.replace({1: "Male", 2: "Female"})
```

```python
model=sm.OLS.from_formula('BPXSY1~RIDAGEYR+BMXBMI+RIAGENDRx',
                          data=df)
result=model.fit()
result.summary()
```
