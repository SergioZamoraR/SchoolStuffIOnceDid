```python
df=pd.read_csv('datasets/antropometria-dataset.csv')
df.head(2)
```

```python
import seaborn as sns
```

```python
sns.boxplot(data=df.loc[:,'cintura'])
```

```python
pivot1=pd.pivot_table(df,
               index='desc_ent',
                       columns='sexo',
                       values='peso',
                       aggfunc='median')
pivot1
```

```python
sns.boxplot(data=pivot1)
```

```python
pivot2=pd.pivot_table(df,
               index='desc_ent',
                       columns='sexo',
                       values='cadera')
pivot2
```

```python
sns.lineplot(data=pivot2)
```

```python
pivot3=pd.pivot_table(df,
               index='desc_ent',
                       columns='sexo',
                       values='sistol',
                       aggfunc='sum')
pivot3
```
